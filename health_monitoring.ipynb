```python
import numpy as np
```

```python
#numpy.random.normal(loc=0.0, scale=1.0, size=None)
#loc = mean center of deviation,scale= standard,size= size of matrix
#std = normal range width (range ~ 4*std)
#heart beat range = 100 ~ 60 
#range = 100-60=40 
#std = 40 / 4 = 10 . As we taking the approximate value it can vary from 8 - 10, for heart beat
#std = 40 / 4 = 10.As we are taking the approximate value it can take 10
```

```python
n = 25
heart_beat = np.random.normal(80,8,n)
blood_pressure = np.random.normal(120,10,n)
body_temperature = np.random.normal(36.8,0.4,n)
oxygen_saturation = np.random.normal(96,1,n)
data = np.stack((heart_beat,
                        blood_pressure,
                        body_temperature,
                        oxygen_saturation))
```

```python
min_val = data.min(axis = 0)
max_val = data.max(axis = 0)
normal_physiology = (data - min_val) / (max_val - min_val)
print(normal_physiology)
```

```python
mean = np.mean(data, axis =0)
std = np.std(data,axis = 0)
median = np.std(data,axis=0)
print(mean)
print(median)
print(std)
```

```python
Z_scores = np.abs((data - mean) /std)
Z_outliers = Z_scores > 3
print(Z_scores)
print(Z_outliers)
```

```python
q1 = np.percentile(data , 25 ,axis =0)
q3 = np.percentile(data, 75 , axis = 0)
Interquartile_range = q3 - q1
lower_bound = q1 - 1.5 * Interquartile_range 
upper_bound = q1 + 1.5 * Interquartile_range 
iqr_outlier = (data < lower_bound) |( data > upper_bound)
print(iqr_outlier)
print(Interquartile_range)
```

```python
ab_mask = Z_outliers | iqr_outlier
ab_val = data[ab_mask.any(axis = 1)]
print(ab_mask)
print(ab_val)
```

```python
patient_status = np.where((heart_beat < 60) | (heart_beat > 100)|(blood_pressure < 90)|(blood_pressure > 130)|
                          (body_temperature < 36.1)|(body_temperature > 37.5)|(oxygen_saturation < 98),"Abnormal","Normal")
print(patient_status)
```
