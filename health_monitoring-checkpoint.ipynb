```python
import numpy as np
```

```python
#numpy.random.normal(loc=0.0, scale=1.0, size=None)
#loc = mean center of deviation,scale= standard,size= size of matrix
#std = normal range width (range ~ 4*std)
#heart beat range = 100 ~ 60 
#range = 100-60=40 
#std = 40 / 4 = 10 . As we taking the approximate value it can vary from 8 - 10, for heart beat
#std = 40 / 4 = 10.As we are taking the approximate value it can take 10, for blood_pressure
```

```python
n = 25
heart_beat = np.random.normal(80,8,n)
blood_pressure = np.random.normal(120,10,n)
body_temperature = np.random.normal(36.8,0.4,n)
oxygen_saturation = np.random.normal(96,1,n)
data = np.stack((heart_beat,
                        blood_pressure,
                        body_temperature,
                        oxygen_saturation))
```

# Normalize physiological singals 

```python
 #normalize = (data - min) / (max - min)
```

```python
min_val = data.min(axis = 0)
max_val = data.max(axis = 0)
normal_physiology = (data - min_val) / (max_val - min_val)
print(normal_physiology)
```

## Performing statistical analysis of Vital parameters 

```python
mean = np.mean(data, axis =0)
std = np.std(data,axis = 0)
median = np.std(data,axis=0)
print(mean)
print(median)
print(std)
```

## Z_scores 

```python
#Z-score: It tells how many standard deviations a data point is away from the mean.
#IQR (Interquartile Range): It measures the spread of the middle 50% of data by subtracting Q1 from Q3
#A Z-score tells how many standard deviations a value is away from the mean.
#In a normal distribution, about 99.7% of data lies within ±3 standard deviations of the mean
#If Z-score ≤ 3 → the value is considered normal
#If Z-score > 3 → the value is very rare / extreme
#If a value > 3 → True (it is an outlier)
#If a value ≤ 3 → False (it’s not an outlier)
```

```python
Z_scores = np.abs((data - mean) /std)
Z_outliers = Z_scores > 3
print(Z_scores)
print(Z_outliers)
```

## IQR

```python
#Quartiles divide a sorted dataset into four equal parts (Q1, Q2/median, Q3), and the Interquartile Range (IQR = Q3 − Q1) measures the spread of the middle 50% of data,
#making it useful for understanding variability and detecting outliers while being less affected by extreme values.
```

```python
q1 = np.percentile(data , 25 ,axis =0)
q3 = np.percentile(data, 75 , axis = 0)
Interquartile_range = q3 - q1
lower_bound = q1 - 1.5 * Interquartile_range 
upper_bound = q1 + 1.5 * Interquartile_range 
iqr_outlier = (data < lower_bound) |( data > upper_bound)
print(iqr_outlier)
print(Interquartile_range)
```

```python
ab_mask = Z_outliers | iqr_outlier
ab_val = data[ab_mask.any(axis = 1)]
print(ab_mask)
print(ab_val)
```
